```python
# To test easyocr with game pictures as input


import easyocr
import cv2
import matplotlib.pyplot as plt
import json
```

```python
# Load and display the source image
img_path = './data/test_1.jpg'
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

```python
# # EasyOCR
# reader = easyocr.Reader(['en'], gpu=False)  # no CUDA, only cpu
# results = reader.readtext(image_rgb)

# # Visualize results
# for (bbox, text, prob) in results:
#     print(f"Detected: '{text}' with confidence {prob:.2f}")
```

```python
# Run OCR
reader = easyocr.Reader(['en'], gpu=False)
results = reader.readtext(image_rgb)

# Extract high-confidence text only
texts = [text for (_, text, prob) in results if prob > 0.6]

# Clean helper
def clean_name(name):
    return name.strip('_').replace('|', 'I').replace('AlI', 'All')

# Locate sections
heroes_start = texts.index("Heroes") + 1
villains_start = texts.index("Villains")
heroes_block = texts[heroes_start:villains_start]
villains_block = texts[villains_start + 1:texts.index("Total", villains_start)]
```

```python
# Parse score table
def parse_team_block(block):
    players = []
    for i in range(0, len(block), 5):
        try:
            player = {
                "name": clean_name(block[i]),
                "score": int(block[i+1]),
                "objective": int(block[i+2]),
                "offense": int(block[i+3]),
                "support": int(block[i+4]),
            }
            players.append(player)
        except (IndexError, ValueError):
            continue
    return players

heroes = parse_team_block(heroes_block)
villains = parse_team_block(villains_block)
```

```python
# Parse information summary
try:
    result_index = texts.index("You Lost")
    result = texts[result_index]
    score = f"{texts[result_index + 1]} : {texts[result_index + 2]}"
except:
    result = "Unknown"
    score = "?"
```

```python
# output structure
summary = {
    "result": result,
    "score": score,
    "teams": {
        "Heroes": heroes,
        "Villains": villains
    }
}

# Output
print(json.dumps(summary, indent=2))
```
